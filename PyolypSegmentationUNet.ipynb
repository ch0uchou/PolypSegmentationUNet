## Reserching Data and Method
### Data Understanding
Bộ Dataset bao gồm các ảnh được trích xuất từ video nội soi. TrainDataset chứa các ảnh có polyp khác nhau và binary mask của ảnh đó. TestDataset chứ các ảnh với nhiều loại polyp khác nhau và ground-true của nó để so sánh.

[Link TrainDataset](https://drive.google.com/file/d/1o8OfBvYE6K-EpDyvzsmMPndnUMwb540R/view)

[Link TestDataset](https://drive.google.com/file/d/1lODorfB33jbd-im-qrtUgWnZXxB94F55/)
### UNet
UNet là một mạng thần kinh tích chập hoàn toàn được phát triển bởi Olaf Ronneberger tại Khoa Khoa học Máy tính của Đại học Freiburg, Đức. Nó được phát triển đặc biệt cho phân đoạn hình ảnh y sinh.
U-net là mạng hình chữ 'U' bao gồm 3 phần:
The Contracting/Downsampling Path
Bottleneck
The Expanding/Upsampling Path
